```python
import pandas as pd
session_feature = pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\session_feature.csv')
buy = pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\yoochoose-buys.dat', 
                  header=None,index_col=False, parse_dates=['TimeStamp'],
                  names=['SesID', 'TimeStamp', 'ItemID', 'Price', 'Quantity'],
                  dtype={'Category':str,'ItemID':str,'SesID':str})
buy_item_feature=pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\buy_item_feature.csv')


test_session_feature=pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\test_session_feature.csv')

test_item_feature=pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\test_item_feature.csv')

solution=pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\solution.dat',sep=';',
                     header=None,names=['SesID','purchased_items'])
solution=pd.concat([pd.Series(row['SesID'],row['purchased_items'].split(',')) for _, row in solution.iterrows()]).reset_index()
solution.columns=['ItemID','SesID']
```

# session classifier

```python
#obtain the input X and target Y 
X=session_feature.drop('SesID',axis=1)
Y=session_feature['SesID'].isin(set(buy['SesID'].astype(int))).astype(int)
```

```python
# split X,Y into train and validation with roughly the same class ratio in train and validation
from sklearn.cross_validation import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# using precision-recall curve as the metric for unbalanced data set
# tune hyper parameters with validation data set to obtain the parameters with highest AUC
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import ParameterGrid
from sklearn import metrics
from sklearn.metrics import precision_recall_curve

param_grid = {'n_estimators': [50], 'criterion': ['entropy','gini'],
              'max_features':['log2','sqrt'],'class_weight':['balanced',None,'balanced_subsample'],
              'max_depth':[20,50,None],'min_samples_leaf':[100,50,1]}

best_auc=0
for i in list(ParameterGrid(param_grid)):
    rf=RandomForestClassifier(n_jobs=-1,class_weight=i['class_weight'],n_estimators=i['n_estimators'],
                              criterion=i['criterion'],max_features=i['max_features'],max_depth=i['max_depth'],
                              min_samples_leaf=i['min_samples_leaf'],random_state=42)
    rf.fit(X_train,y_train)
    rf_prob=rf.predict_proba(X_val)[:,1]
    precision, recall, _=precision_recall_curve(y_val,rf_prob)
    auc= metrics.auc(recall,precision)
    if auc>best_auc:
        best_auc=auc
        best_parameter=i
print 'best parameters: ' + str(best_parameter)
print 'Area under curve: '+str(best_auc)
```

#tuning the probability thresholds of the session and item classifiers

```python
#defining the functions to compute the score required in the metric
def purchase_score(session_feature,X_val,y_val,session_prob):
    probs=best_rf.predict_proba(X_val)[:,1]
    predicted_buy_session=session_feature.iloc[X_val.iloc[numpy.where(probs>session_prob)].index]['SesID']
    ground_truth=session_feature.iloc[y_val.iloc[np.where(y_val==1)].index]['SesID']
    score=0.05*(sum(predicted_buy_session.isin(ground_truth))-sum(~predicted_buy_session.isin(ground_truth)))
    correct_predicted_session=predicted_buy_session[predicted_buy_session.isin(ground_truth)]
    return (score,np.array(correct_predicted_session))

def jaccard_score(correct_predicted_session,buy,buy_item_feature,item_prob_threshold):
    if len(correct_predicted_session)==0:
        return 0
    temp=buy[['SesID','ItemID']].apply(lambda x: str(x[0])+'_'+str(x[1]),axis=1)
    item_feature=buy_item_feature[buy_item_feature['SesID'].isin(correct_predicted_session)]
    predict_buy_item=np.array(best_item_rf.predict_proba(item_feature.drop(['ItemID','SesID'],axis=1))[:,1]\
                              >item_prob_threshold,dtype=int)
    ground_truth=np.array(item_feature[['SesID','ItemID']].apply(lambda x: str(x[0])+'_'+str(x[1]),axis=1). \
                          isin(temp),dtype=int)
    table=pd.DataFrame(data={'prediction':predict_buy_item, 'ground_truth':ground_truth,\
                           'SesID':item_feature['SesID']})
    table['union']=table['prediction'] | table['ground_truth']
    table['intersection']=table['prediction'] & table['ground_truth']
    return sum(table.groupby('SesID').sum()['intersection']/table.groupby('SesID').sum()['union'])

def total_score(session_prob,item_prob_threshold,session_feature,X_val,y_val,buy,buy_item_feature):
    purchased_score, purchased_session=purchase_score(session_feature,X_val,y_val,session_prob)
    return purchased_score+jaccard_score(purchased_session,buy,buy_item_feature,item_prob_threshold)
```

```python
#finding the optimal probability thresholds for the session and item classifiers
from scipy.optimize import differential_evolution
prob_threshold=differential_evolution(lambda x: -total_score(x[0],x[1],session_feature,X_val,\
                                                            y_val,buy,buy_item_feature),[(1e-2,0.99),(1e-2,0.99)]).x
```
