```python
import pandas as pd
click = pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\yoochoose-clicks.dat',
                    header=None, index_col=False, parse_dates=['TimeStamp'],
                    names=['SesID', 'TimeStamp','ItemID','Category'])
buy = pd.read_csv(r'C:\Users\Mark simulation\Downloads\New folder\yoochoose-buys.dat', 
                  header=None,index_col=False, parse_dates=['TimeStamp'],
                  names=['SesID', 'TimeStamp', 'ItemID', 'Price', 'Quantity'])
```

# Exploratory analysis

```python
#top 10 sold item
import heapq
top_10_sold=zip(buy.groupby('ItemID').size(),buy.groupby('ItemID').count().index)
heapq.heapify(top_10_sold)
heapq.nlargest(10,top_10_sold)
plt.title("Top 10 sold items")
plt.bar(range(10),[i[0] for i in heapq.nlargest(10,top_10_sold)] , align="center")
plt.xticks(range(10), [i[1] for i in heapq.nlargest(10,top_10_sold)],rotation='vertical')
plt.xlim([-1, 10])
plt.xlabel('item ID')
plt.ylabel('sales')
plt.tight_layout() 
```

```python
# purchase ratio with respect to date
# purchase ratio is defined as: number of unique items bought dividied by the number of unique clicked items
buy_vs_date = buy["TimeStamp"].groupby(buy["TimeStamp"].apply(lambda x : x.date()))
buy_vs_date = buy_vs_date.count()
buy_vs_date = buy_vs_date.to_frame(name="buy")

click_vs_date = click["TimeStamp"].groupby(click["TimeStamp"].apply(lambda x : x.date()))
click_vs_date = click_vs_date.count()
click_vs_date = click_vs_date.to_frame(name="click")

buy_click_vs_date = pd.merge(buy_vs_date, click_vs_date, how='inner', left_index=True, right_index=True)
buy_click_vs_date['ratio']=buy_click_vs_date.apply(lambda x: 1.0*x[0]/x[1],axis=1)
```

```python
buy_click_vs_date['ratio'].plot()
plt.title("Purchase ratio vs date")
plt.ylabel('purchase ratio')
plt.xlabel('Date')
```
